## Classic Summary

```python
data_path = "Results/AMI-Spanish/classic_"
sizes = ["32","64","128","256","512"]
folds=["1","2","3","4","5"]
runs=["1","2","3"]
```

```python
df = pd.read_csv(data_path+"512_1_1_test.csv")
cols = df.columns
cols
```

```python
dfs=[]
```

```python
for size in sizes:
    df1 = pd.DataFrame(columns=cols)
    for fold in folds:
        for run in runs:
            df = pd.read_csv(data_path+size+"_"+fold+"_"+run+"_test.csv")
            df['name']=[val.ljust(3,"0") for val in df['name']]
            df1=pd.concat([df1,df])
    df1.sort_values(by='name',inplace=True)
    df_lr = df1.iloc[:15].describe()
    df_svc = df1.iloc[15:].describe()
    dfs.append(df_lr)
    dfs.append(df_svc)
```

```python
dfs[1].head()
```
